# Rozdělení dat

```python
import os
import random
from ultralytics import YOLO
import torch
```

```python
print("CUDA available:", torch.cuda.is_available())
print("Device count:", torch.cuda.device_count())
print("Current device:", torch.cuda.current_device() if torch.cuda.is_available() else "CPU only")
```

```python
TRAINING_ENABLED = False

TRAINING_EPOCHS = 50
IMAGE_SIZE = 640
BATCH_SIZE = 16
WORKERS = 2
DEVICE = '0' if os.getenv('CUDA_AVAILABLE', '0') == '1' else 'cpu'
MODEL = 'yolov8s.pt'
PATIENCE = 15
BATCH_SIZE = 16
WORKERS = 8

PROJECT_DIR = os.path.dirname(os.path.abspath("."))
NAME = MODEL.split('.')[0]
DATA_DIR = os.path.join(PROJECT_DIR, '..', 'data')

print(f'Project directory: {PROJECT_DIR}')
print(f'Data directory: {DATA_DIR}')
print(f'Using device: {DEVICE}')
```

```python
def train_yolo(yaml_path=None):

    model = YOLO(MODEL)
    print(f'Starting training YOLO model {MODEL} on device {DEVICE} with yaml path {yaml_path}...')

    return model.train(
        data=yaml_path,
        epochs=TRAINING_EPOCHS,
        patience=PATIENCE,
        imgsz=IMAGE_SIZE,
        batch=BATCH_SIZE,
        workers=WORKERS,
        project="../models",
        name=NAME,
        plots=True,
        save=True, 
        cos_lr=True,
    )
```

```python
yaml_path = os.path.join(os.getcwd(), "..", "data", "data.yaml")
print(f'Trénink s daty z: {yaml_path}')
```

```python
if TRAINING_ENABLED:
    train_yolo(yaml_path)
else:
    print("Trénink vypnutý, přeskočeno.")
```
